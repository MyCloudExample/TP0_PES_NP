```python
#1. Generar seal cuadrada, triangular, delta dirac en python si nel uso de librerias
#Parte resuleta por el docente

import matplotlib.pyplot as plt

# Seal cuadrada (escaln)
senial_cuad = [] #inicializo un array vaco
for a in range(5):
    senial_cuad.append(0)
for a in range(5):
    senial_cuad.append(1)
print(senial_cuad)
plt.title("Escaln")
plt.plot(senial_cuad)
plt.show()

# Seal delta de dirac
delta_dirac = []
for a in range(100):
    delta_dirac.append(0)
dd= int(len(delta_dirac)/2) #obtener la posicin del medio del array
delta_dirac[dd] = 1
print("el largo del vector delta_dirac es: {}".format(len(delta_dirac)))
plt.title("Dirac")
plt.plot(delta_dirac)
plt.show()


# Seal triangular
senial_tria = []
for a in range(5):
    senial_tria.append(a)
for a in range(5,0,-1):
    senial_tria.append(a)
plt.title("Triangular")
plt.plot(senial_tria)
plt.show()
```
